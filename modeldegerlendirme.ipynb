```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc, f1_score
import seaborn as sns
```

```python
# Gerçek ve tahmin edilen değerler (örnek veri)
y_true = np.array([0, 1, 1, 0, 1, 0, 1, 0, 0, 1])  # Gerçek etiketler
y_pred = np.array([0, 1, 0, 0, 1, 0, 1, 1, 0, 1])  # Modelin tahminleri
y_prob = np.array([0.1, 0.9, 0.3, 0.2, 0.8, 0.1, 0.9, 0.7, 0.2, 0.85])  # Modelin olasılık tahminleri
```

```python
# Confusion Matrix
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(5, 4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["Negatif", "Pozitif"], yticklabels=["Negatif", "Pozitif"])
plt.xlabel("Tahmin Edilen")
plt.ylabel("Gerçek")
plt.title("Confusion Matrix")
plt.show()
```

```python
# ROC-AUC Skoru
fpr, tpr, _ = roc_curve(y_true, y_prob)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(6, 5))
plt.plot(fpr, tpr, color="blue", lw=2, label=f"ROC curve (AUC = {roc_auc:.2f})")
plt.plot([0, 1], [0, 1], color="gray", linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```

```python
# F1-Score
f1 = f1_score(y_true, y_pred)
print(f"F1-Score: {f1:.2f}")

# Classification Report
print("Classification Report:\n", classification_report(y_true, y_pred))
```
